```python
import os
from pathlib import Path
from glob import glob
import nrrd #pip install pynrrd, if pynrrd is not already installed
import nibabel as nib #pip install |nibabel, if nibabel is not already installed
import numpy as np
from scipy.ndimage import zoom
```

```python
folder_path = r"C:\Users\BRISHTI\Documents\images dong"
file_list = os.listdir(folder_path)

outDir1 = r"C:\Users\BRISHTI\Documents\org data\converted imgs"

for file in file_list:
    file_path = os.path.join(folder_path, file)
    
    # Load nrrd
    _nrrd = nrrd.read(file_path)
    data = _nrrd[0]
    header = _nrrd[1]
    
    # Save nifti
    img = nib.Nifti1Image(data, np.eye(4))
    filename = Path(file).stem
    filename += '-img.nii.gz'
    outpath = os.path.join(outDir1, filename)
    nib.save(img, outpath)
```

```python
#entire directory for (256,256,64)
list1 = glob(r"C:\Users\BRISHTI\Documents\org data\converted imgs\*.nii.gz") 
list2= glob(r"C:\Users\BRISHTI\Documents\org data\converted masks\*.nii.gz")
for x,y in zip(list1,list2):
    img = nib.load(x)
    mask= nib.load(y)
    img_rib = np.array(img.dataobj)
    mask_rib= np.array(mask.dataobj)
    print('before conversion'+str(mask_rib.shape))
    den_0=mask_rib.shape[0]
    den_1=mask_rib.shape[1]
    den_2=mask_rib.shape[2]   
    b=(img.affine[0][0])
    c=(img.affine[1][1])
    d=(img.affine[2][2])
    print(b,c,d)
    a=np.min(img_rib)
    b1=np.min(mask_rib)
    img_array = zoom(img_rib, ((256/den_0),(256/den_1),(64/den_2)),cval=a)
    mask_array=zoom(mask_rib, ((256/den_0),(256/den_1),(64/den_2)),order=0,mode='nearest',cval=b1)
    print('after conversion'+str(mask_array.shape))
    img_affine=nib.affines.rescale_affine(img.affine,(256,256,64),((den_0/256),(den_1/256),(den_2/64)))
    mask_affine=nib.affines.rescale_affine(mask.affine,(256,256,64),((den_0/256),(den_1/256),(den_2/64)))
    ni_img = nib.Nifti1Image(img_array, img_affine)
    ni_mask= nib.Nifti1Image(mask_array,mask_affine)
    filename1 = Path(x).stem
    filename1 = Path(filename1).stem
    filename1 += '.nii.gz'
    filename2 = Path(y).stem
    filename2 = Path(filename2).stem
    filename2 += '.nii.gz'
    outDir1 =r"C:\Users\BRISHTI\Documents\org data\resized imgs(256)"
    outDir2=r"C:\Users\BRISHTI\Documents\org data\resized masks(256)"
    outpath1 = os.path.join(outDir1, filename1)
    outpath2= os.path.join(outDir2, filename2)
    print(outpath1)
    print(outpath2)
    nib.save(ni_img, outpath1)
    nib.save(ni_mask,outpath2)
```
